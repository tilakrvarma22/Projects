###1.LinearRegression
* It is algorithm in which feature and label are linearly dependent to each other.
* In this a independt variable is linearly dependent on dependent variable.
* `y = bias + weight*X`

 where **bias** - it is factor that ensures that output must not too big and not too small.

 **weight** - it is slope of the function.It controls on how strongly would feature will influence the output.

```python
# Creating a model
from sklearn.linear_model import LinearRegression
```

```python
model_1=LinearRegression()
```

```python
model_1.fit(X_train,y_train)
```

```python
# Prediction of outputs
y_train_predict=model_1.predict(X_train)
y_test_predict=model_1.predict(X_test)
```

### Performance Metrics
* Performance metrics will be the factor on which the model performace will be made.
* It track and process of  the model
* Performace metrics are:
  1. Mean Absolute Error
  2. Root Mean Square error
  3. Mean Square Error
  4. Accuracy
  5. Precision
  6. Recall
  7. Confusion Metrics
  8. F1-Score
* This problem is an regression problem.
So the performace metrics will be determined by the loss function of the model.
* Metrics used are:
  1. MAE
  2. MSE

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
train_loss=mae(y_train,y_train_predict)
test_loss=mae(y_test,y_test_predict)
```

* Ensemble learning is based on the 2 main concepts which are:
   1. **Bagging** - it is fitting a different model on different datasets and then averaging thier predictions or outputs.Its aims in enhancing performace execution and precisio of algorithm.
   2. **Boosting** - it trains the model sequentially focus on the error made the previous model.It iteratively modifies the weight of the observation based on the last classification.
