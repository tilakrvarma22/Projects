###1.Decision Tree

```python

#Importing
from sklearn.tree import DecisionTreeRegressor

# Defining a model
model_1=DecisionTreeRegressor(random_state=1)

# Fitting a model
model_1.fit(X_train,y_train)

#Prdicting results
y_val_predict=model_1.predict(X_val)
y_train_predict=model_1.predict(X_train)
```

#### Evaluation metrics
For the problems like these we metrics like
* MAE(Mean Abolute Error)
* MSE(Mean Square Error)
* RSME(Root mean squraed error)

```python
#importing
from sklearn.metrics import mean_absolute_error
train_loss_1=mean_absolute_error(y_train,y_train_predict)
val_loss_1=mean_absolute_error(y_val,y_val_predict)
```

###2.Random Forest

Random forest : It is ensemble learning algorithm in which mulitple decision trees are created and take the avg of predictions made out of it.
It better than decision tree and also performs better than decision tree


```python
## importing
from sklearn.ensemble import RandomForestRegressor

# Craeting a model
model_2=RandomForestRegressor(random_state=1)

# Fitting a model
model_2.fit(X_train,y_train)

# predict
y_train_predict_2=model_2.predict(X_train)
y_val_predict_2=model_2.predict(X_val)
```

```python
#loss function
train_loss_2=mean_absolute_error(y_train,y_train_predict_2)
val_loss_2=mean_absolute_error(y_val,y_val_predict_2)
```

* **Training loss of Decision tree** is less than **Training loss of random forest** Bcoz a decision tree combines some decisions but random forest has to combine various trees decisions
* But **Validation loss of decision tree** is more than **validation loss of random forest** bzoz in random forest we take the avg of various decision trees this leads to better generalisation and lower validation loss
* **Generalisation**:It is a  context in ml which states that how model performs on unseen data.

###3.Lasso Regression
It is also called **L1-Regularisation**. Its adds a penalty term to the cost function.
It is type of linear regression that uses shrinkage to improves accuracy of the model.

**Regularisation**: It prevents the model from overfitting by adding extra information to it.


```python
## import
from sklearn.linear_model import Lasso

## Creation of model
model_3=Lasso(random_state=1)

#fitting a model
model_3.fit(X_train,y_train)

#predictions
y_train_predict_3=model_3.predict(X_train)
y_val_predict_3=model_3.predict(X_val)
```

```python
#Loss function
train_loss_3=mean_absolute_error(y_train,y_train_predict_3)
val_loss_3=mean_absolute_error(y_val,y_val_predict_3)
```
